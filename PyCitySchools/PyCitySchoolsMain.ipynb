```python
import pandas as pd
```

```python
school_file = '../Resources/schools_complete.csv'
student_file = '../Resources/students_complete.csv'

school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)
```

```python
copy_school_sum = school_df.copy()
```

## District Summary

```python
total_schools = school_df["school_name"].count()
total_students = student_df["school_name"].count()
total_budget = school_df["budget"].sum()

avg_math_score = student_df["math_score"].mean()
avg_read_score = student_df["reading_score"].mean()

math_pass = student_df.loc[(student_df["math_score"] >= 70)]
count_pass_math = math_pass["math_score"].count()

per_math_pass = (count_pass_math/total_students)*100

read_pass = student_df.loc[(student_df["reading_score"] >= 70)]
count_pass_read = read_pass["reading_score"].count()

per_read_pass = (count_pass_read/total_students)*100

overall_pass = (per_math_pass + per_read_pass)/2
```

```python
district_summary = {"Total Schools" : total_schools,
                   "Total Students" : total_students,
                   "Total Budget" : total_budget,
                   "Average Math Score" : avg_math_score,
                   "Average Reading Score" : avg_read_score,
                    "% Passing Math" : per_math_pass,
                   "% Passing Reading" : per_read_pass,
                   "% Overall Passing" : overall_pass}


district_summary_df = pd.DataFrame([district_summary])                                              
                                   
district_summary_df = district_summary_df[["Total Schools",
                   "Total Students","Total Budget","Average Math Score", "Average Reading Score",
                   "% Passing Math", "% Passing Reading", "% Overall Passing" ]]
district_summary_df
```

# School Summary

```python
del copy_school_sum['School ID']
```

```python
copy_school_sum['Per Student Budget'] = copy_school_sum['budget']/copy_school_sum['size']
```

```python
avg_math_read_tbl = student_df.groupby(['school_name'])['reading_score', 'math_score'].mean().reset_index()
```

```python
copy_school_sum = copy_school_sum.merge(avg_math_read_tbl, on='school_name', how="outer")
```

```python
summary_passing_read = student_df[student_df['reading_score']>=70]
summary_passing_math = student_df[student_df['math_score']>=70]
```

```python
pass_read_count_sum = summary_passing_read.groupby(["school_name"])['reading_score'].count().reset_index()
pass_read_count_sum.rename({'reading_score' : 'Reading Count'}, axis=1, inplace= True)
```

```python
pass_math_count_sum = summary_passing_math.groupby(["school_name"])['math_score'].count().reset_index()
pass_math_count_sum.rename({'math_score' : 'Math Count'}, axis=1, inplace=True)
```

```python
pass_count = pass_math_count_sum.merge(pass_read_count_sum, on="school_name", how='inner')
```

```python
copy_school_sum = copy_school_sum.merge(pass_count, on="school_name", how='outer')
```

```python
copy_school_sum['% Passing Math'] = (copy_school_sum['Math Count']/copy_school_sum['size'])*100

copy_school_sum['% Passing Reading'] = (copy_school_sum['Reading Count']/copy_school_sum['size'])*100
```

```python
del copy_school_sum['Math Count']
del copy_school_sum['Reading Count']
```

```python
copy_school_sum['% Overall Passing'] = (copy_school_sum['% Passing Math'] + copy_school_sum['% Passing Reading'])/2
```

```python
copy_school_sum.rename({'reading_score':'Avg. Reading Score',
                             'math_score':'Avg. Math Score'}, axis=1, inplace=True)
copy_school_sum
```

##  Scores by School Spending

```python
scores_by_school_spending = copy_school_sum.copy()
```

```python
bins = [0, 585, 630, 645, 680]
spending_labels = ['<585', '$585-630', '$630-645', '$645-680']
```

```python
bins_school_spending = pd.cut(scores_by_school_spending['Per Student Budget'], bins, labels=spending_labels)

bins_school_spending = pd.DataFrame(bins_school_spending)
 
copy_school_sum['Spending Level'] = bins_school_spending
```

```python
scores_by_school_spending = copy_school_sum.groupby(['Spending Level'])['Avg. Math Score','Avg. Reading Score',
                                                      '% Passing Math','% Passing Reading','% Overall Passing'].mean()
```
